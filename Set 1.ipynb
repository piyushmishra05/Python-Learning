# OPERATORS IN PYTHON

Operators are special symbols or words that perform operations on variables and values

Types of operators:-

1) Arithematic operators:- It is used to perform basic mathematical operations like addition, subtraction, multiplication, etc.

Operator Name           Example      Description
'+' Addition            'a+b'        Adds two numbers
'-' Subtraction         'a-b'        Subtracts second from first
'*' Multiplication      'a*b'        Multiplies two numbers
'/' Division            'a/b'        Divides and returns float result
'//' Floor Division     'a//b'       Divides and removes decimals
  
2) Assignment operators:- Are used to assign values to variable or to update them using operations
Operator Name                        Example      Meaning
'=' Assigns value                    'x = 5'      Assigns 5 to 'x'
'+=' Add and assign                  'x += 2'     'x = x + 2'
'-=' Subtract and assign             'x -= 2'     'x = x - 2'
'*=' Multiply and assign             'x *= 2'     'x = x * 2'
'/=' Divide and assign               'x /= 2'     'x = x / 2'
'//=' Floor Divide and assign        'x //= 2'    'x = x // 2'



3) Comparison operators:- Used to compare two values. They always return True or False
Operator Name                 Example      
'==' Equal to                  'a==b'       == Used to compare two values and = used to assign values
'!' Not equal to               'a!=b'        
'>' Greater Than               'a>b'        
'<' Less Than                  'a<b'        
'>=' Greater than Equal to     'a>=b'
'<=' Less than Equal to        'a<=b'



4) Logical operators:- Used to combine multiple conditions. They return True or False based on logic
Operator Name         Meaning                                               Example
'and'                 True if **both** conditions are True                  'x > 5' and x < 10'
'or'                  True If **at least one** condition is True            'x < 5' or x > 20'
'not'                 Reverses the result (True -> False, False -> True)    'not(x > 5)'


5) Membership operators:-
Operator Name       Meaning      
'in'                True if value exists                 '"a" in "apple"'       
'not in'            True if value **does not** exist     '5 not in [1,2,3]        


6) Identity operators

# LOOPS  (Iterations)
Loops are programming constructs which are used to implement same logic n number of times(multiple times).
There are Two types of loops:- 
1) **For Loop(Iterative Loop)**:- It iterates all the elements in the sequence or collection and implement the logic. It will terminate after all the elements in the sequence are completed. It is one of the widely used loop.(starts from first and end on the last)

 
2)**While Loop(Conditional Loop)**:- It will execute only till the condition is true. It will terminate once the condition turns out to be False.
Terminology used: In while loop initializer is used(From where to start)
Increment and Decrement plays an important role
Condition is necessary to give in while loop

Feature           "For" Loop                          "While" Loop

Use Case          Known number of iterations           Unknown or conditional iterations
Syntax Focus      Iterates over sequence or range      Runs while a condition is "True"
Control type      Count-Controlled                     Condition-Controlled
Example           "for i in range(5):"                 "while i < 5:"   # "i" is the iterator here

**Initializer** is used to give instruction to the system on where to start the iteration.
**Iterator** is used for executing a set of instruction multiple times

```python
count = 0 #INITIALIZER

while (count <= 10):
    count += 1   #pre-increment
    print(count)
```

# Loop Statements 
There are two types of loop statements:- 
Break: This statement is used to exit the loop early without completing the iteration mentioned in the range.
Continue: This statement skips the current loop iteration based on the condition given by the user and moves to the next one.

```python
for i in range(1,10):
    if i == 5:
        break
    print("Number is", i)
print("Out of loop")
```

```python
i = 1
while i < 10:
    if i == 5:
        break
    print("Number is", i)
    i += 1
print("Out of loop")
```

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
```

```python
# For range 1 to 60 skip even numbers and print out of loop - using for loop
for i in range(1,60):
    if i % 2 == 0:
        continue
    print(i,"Is an Odd number")
print("Out of loop")
```

                         6) List supports Item Assignment. (To change value)

**Properties of Sets:-** 1) Set allows mixed data type.

                         4) Sets are Mutable in nature.
                         5) Set donot support Indexing or Slicing.
                         6) Set donot allow Item Assignment.
                         7) Set allow us to perform set operations such as Union, Intersection, Difference

                          6) Tuple Donot allow Item Assignment.

# Indexing and Slicing
**Indexing:-** 1) We use indexing to assign an index value to the string
2) We use square brackets[ ] to get the index value and it has an Start and End value [Start = Inclusive, End = Exclusive].
Start value is Inclusive which means whatever index number you give as the start value that index will be fetched out in the output.
End value is Exclusive in slicing it will stop the indexing just before the end value.
3) In Indexing , the index value always starts with 0 if it is a negative indexing it should always start with -1.

**Slicing:-** 1) If you want to fetch out only any given index number we can use Slicing.
2) When using negative slicing if you mention a smaller value at the start point and comparitively bigger value at the end point, it will throw an error.
3) In slicing if you don't mention any start value the system by default takes the first index value is the default value.
4) If you don't assign any end value the system by default takes the last index number as the end value.

# Tuples are faster than lists because of their immutability

Reasons:
Fixed Size:
Tuples are immutable, so python can allocate a fixed amount of memory and optimize access.

Less Overhead:
Lists have extra methods to support modifications (like append, extend, etc). Tuples don't need this, so they're lighter.

Use tuples when your data shouldn't change and performance matters
Use lists when your data needs to be modified.

```python
# Set operations
# Creating list

A = [90, 80, 70, 60, 5, 4, 3, 2, 1]
B = [55, 3, 76, 6, 22, 1, 99, 5]

# coverting list into set
setA = set(A)
print(setA)

setB = set(B)
print(setB)
```

```python
# Union - Combines all the unique elements from the set
A = [90, 80, 70, 60, 5, 4, 3, 2, 1]
B = [55, 3, 76, 6, 22, 1, 99, 5]

print(setA.union(setB))  # 2 ways to write, works as or value either in A or B
print(setA | setB)
```

```python
# Using Intersection:- Shows only the similar value from both the Sets

A = [90, 80, 70, 60, 5, 4, 3, 2, 1]
B = [55, 3, 70, 6, 22, 1, 90, 5]

setA = set(A)
setB = set(B)

print(setA.intersection(setB))  # 2 Ways to write
print(setA & setB)
```

```python
# Difference:- All the elements of 1 set & matching from the rest
# Elements present in one set but not in another
A = [90, 80, 70, 60, 5, 4, 3, 2, 1]
B = [55, 3, 70, 6, 22, 1, 90, 5]

setA = set(A)
setB = set(B)

print(setB.difference(setA))  # So only the unique items of setB are returned.use(-)
print(setA-setB) # value in A which are not present in B
```

# Copy
.copy() creates a separate dictionary object with the same Key value pairs. 
Changes to the copy won't affect the original
As undo is not allowed and you dont like the changes

```python
# copy
book = {
    "title": "The Alchemist",
    "author": "Paulo Coelho",
    "year": 1998
}

# Copying the book dictionary
book_copy = book.copy()

# Modify the copy
book_copy["year"] = 2023

print("Original Dictionary:", book)
print("Copied Dictionary:", book_copy)
```
