# Difference between Array and List
1) Arrays are homogeneous and **list are heterogeneous**
2) Arrays are Multidimensional and **list are one dimensional**
3) Array stores data in Contiguous memory and **list doesn't store data in contiguous memory**
4) Arrays are faster in comparision to list
5) List is slower as compared to arrays
6) Array follows SIMD- Single instruction multiple data
7) List Doesn't follow SIMD

# Allocating Size

When we give a size like int32, float64, etc.
We are telling python how much memory to use and how accurate the numbers should be

Why we do it:
Save memory - smaller size - use less space
Faster - smaller data is quicker to process
Better control - we choose how exact we want the numbers
Useful when working with big data or ML models

```python
a3 = np.array([34,45,901,56.78,21.26,-102], dtype ="int64")  # datatype
print(a3)


# **dtype**
# It tells Numpy what type of data to store in the array
# All elements are converted to 64-bit integers
# So decimal parts (.78, .26) will be truncated (not rounded)
```

# SIMD stands for Single Instruction, Multiple Data.

SIMD One instruction, applied to many data points at once.

Used in NumPy for fast operations.

Helps in faster computation, especially for large arrays

```python
import numpy as np

# without SIMD (regular Python)

data = [1, 2, 3, 4]

result = []

for x in data:
    result.append(x + 5)

print("Without SIMD:", result)
```

```python
# with SIMD (NumPy does it internally)

arr = np.array([1, 2, 3, 4])
result_simd = arr + 5

print("With SIMD:", result_simd)
```

# Reshape

It is primarily refers to changing the dimensions of a data structure, without altering its underlying data.

```python
import numpy as np

# Create a 1D NumPy array
arr = np.array([1, 2, 3, 4, 5, 6])

# Reshape the 1D array into a 2x3 matrix
reshaped_arr = arr.reshape((2, 3))
print("Reshaped 2x3 array:")
print(reshaped_arr)

# Reshape the array into a 3x2 matrix using -1 for an unknown dimension
reshaped_arr_auto = arr.reshape((3, -1))
print("\nReshaped 3x2 array (with -1):")
print(reshaped_arr_auto)
```

# Aggregate function

np.min(a)

np.max(a)

np.sum(a)

np.mean(a)

np.prod(a)

np.medium(a)

np.std(a)

np.var(a)

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Aggregations over the entire array
print(f"Minimum value: {np.min(arr)}")
print(f"Maximum value: {np.max(arr)}")
print(f"Sum of elements: {np.sum(arr)}")
print(f"Mean of elements: {np.mean(arr)}")
print(f"Product of elements: {np.prod(arr)}")
print(f"Median of elements: {np.median(arr)}")
print(f"Standard deviation: {np.std(arr)}")    # while standard deviation is the square root of the variance
print(f"Variance: {np.var(arr)}")             # variance represents the average of the squared differences from the mean

# Aggregations along a specific axis (e.g., axis=0 for columns)
print(f"Column-wise sum: {np.sum(arr, axis=0)}")
```

# List Comprehension

List comprehension is a shorter way to create a new list by applying an operation or filter on an existing iterable(like a list or range)

```python
flowers = ["rose", "lily", "tulip", "marigold", "sunflower"]
# we have a list of flower names and we want to convert them to uppercase
```

```python
# We have a list of flower names we want to convert them to uppercase
# Traditional way of creating a list

flowersList = []
for i in flowers:
    flowersList.append(i.upper())
    print(flowersList)
```

```python
# Using LC
FlowersList = [i.upper() for i in flowers]
print(FlowersList)
```

```python
# Create a list of square of numbers from 1 to 5
```

```python
squares = []
for x in range(1, 6):
    squares.append(x**2)

print(squares)
    
```

```python
squares = [x**2 for x in range(1, 6)]
print(squares)
```

```python
# Add 10 to each number in nums = [1,2,3,4] solve using with or without List comprehension
```

```python
# Without LC
nums = [1, 2, 3, 4]
result = []
for x in nums:
    result.append(x + 10)

print(result)
```

```python
# With LC
nums = [1, 2, 3, 4]
result = [x + 10 for x in nums]
print(result)
```

```python
# Get only even numbers in list: [1, 2, 3, 4, 5, 6] solve using with or without List comprehension
```

```python
# With LC
nums = [1, 2, 3, 4, 5, 6]
evens = [x for x in nums if x % 2 == 0]
print(evens)
```

```python
# Without LC
nums = [1, 2, 3, 4, 5, 6]
evens = []

for x in nums:
    if x % 2 == 0:
        evens.append(x)

print(evens)
```

```python
# LC with if-else
# Syntax

# [expression_if_true if condition else expression_if_false for item in iterable]
```

```python
nums = [1, 2, 3, 4, 5]
result = []

for x in nums:
    if x % 2 == 0:
        result.append(f"{x}: Even")
    else:
        result.append(f"{x}: Odd")

print(result)
```

```python
nums = [1, 2, 3, 4, 5]
result = [f"{x}: Even" if x % 2 == 0 else f"{x}: Odd" for x in nums]
print(result)
```

```python
# if score >= 40 then pass, else fail solve using with or without list comprehension
```

```python
scores = [25, 67, 40, 90, 39]
results = [f"{score}: Pass" if score >= 40 else f"{score}: Fail" for score in scores]
print(results)
```

```python
scores = [25, 67, 40, 90, 39]
results = []

for score in scores:
    if score >= 40:
        results.append(f"{score}: Pass")
    else:
        results.append(f"{score}: Fail")

print(results)
```
