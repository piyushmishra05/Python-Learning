# Loc and iloc

**df.loc = location specific
  df.iloc = index location specific

- For accessing a DataFrame or fetching a value and set of value with respect to their index,
  we use either loc or iloc


**loc:-** loc take row label for accessing the data
Syntax loc - [start(inclusive):end(inclusive)]

**iloc:-** iloc will only take indexing for accessing the data

```python
# Select a single rows by labels

datal.loc["E2"]   # Get Mehvish's Row
```

```python
# Select multiple rows by labels

datal.loc[["E1", "E3", "E5"]]   # Get Piyush, Neer and Ganesh
```

```python
# Select specific rows and columns by label

datal.loc[["E1", "E3"], ["Employee","Salary"]]
```

```python
# Slice by label range

datal.loc["E2":"E4", "Department":"Salary"]
```

```python
# Single row by position using iloc

datal.iloc[0:3]   # Row at position 1 (Mehvish) (inclusive)

# Equivalent loc: datal.loc["E2"]
```

```python
datal.iloc[[1]]
# 1 bracket [] (Series)

# 2 brackets [[]]  Table (DataFrame)
```

```python
# Multiple Rows by positions
datal.iloc[[0,3]]  # Rows at positions 0 and 3

# Equivalent loc: datal.loc[["E1", "E4"]]
```

```python
# Row Range (Slice)
datal.iloc[1:4]  # Rows at positions 1, 2, 3
# Equivalent loc: datal.loc["E2": "E4"]
```

```python
datal
```

```python
# Single cell (Row & Column by position)

print(datal.iloc[2,3])   # Row 2, Column 3 salary of Neer

# Equivalent loc: datal.loc["E3", "Salary"]
```

```python
# Subset of Rows & Columns

print(datal.iloc[0:3, 0:2])   # First 3 rows, first 2 columns

# Equivalent loc: datal.loc["E1":"E3", ["Employee", "Department"]]
```

# Delete - del

- Removes a column permanently from the DataFrame (in-place)

- No return value - it directly modifies the DataFrame

- Works only for columns, not rows

Syntax: del df["ColumnName"]

# Drop

- Can remove rows or columns

- Returns a new DataFrame by default (Unless inplace = True)

- Needs axis or index / columns parameter

Syntax: 

Drop column: df.drop(columns = "Score")

Drop row: df.drop(index = "E5")

```python
del datal["RevisedScore"]   # Removes column permanently - if you run again will get error
datal
```

```python
datal.drop("E4")   # By default will take row
# E4-removed
```

```python
# But if we run df again the dropped value will be visible again
datal
```

```python
# If trying to use drop with column will get error
#datal.drop("Score")   # Key Error: "["Score"] not found in axis"
# We need to specify axis
```

```python
# With axis it will not give error

datal = datal.drop("Score", axis = 1)
datal
```

```python
# Using inplace = True will permanently remove
datal.drop("City", axis = 1, inplace = True)
datal
```

```python
# We will use ".index" to solve this problem

datal.drop(datal[(datal["Department"] == "IT") & (datal["Age"] > 25)].index, inplace = True)
datal
# & - AND Operator
# | - OR Operator
```

**Q2:- Calculate the average marks and add as a new column

**Q3:- Fetch all the students from Tier 1

**Q6:- Using index Fetch details of the students in 2nd Row

**Q8:- Fetch Rows from index 0 to 2 and Column index 1 to 4

**Q9:- Fetch Name and tier for every alternate student with labels

**Q10:- Fetch name and average marks for students in reverse order

```python
# Create a DataFrame named df_students for 6 students with the following columns:
# StudentID Name SQL_Marks Python_Marks Excel_Marks Address Tier

import pandas as pd

data = {
    "StudentID": [101, 102, 103, 104, 105, 106],
    "Name": ["Piyush", "Mehvish", "Neer", "Prarthana", "Ganesh", "Abhishek"],
    "SQL_Marks": [91, 85, 92, 74, 88, 80],
    "Python_Marks": [92, 90, 85, 70, 89, 70],
    "Excel_Marks": [93, 80, 90, 72, 85, 71],
    "Address": ["Pune", "Mumbai", "Pune", "Jhansi", "Bhusawal", "Khopoli"],
    "Tier": ["Tier 1", "Tier 1", "Tier 2", "Tier 2", "Tier 3", "Tier 3"]
}

df_students = pd.DataFrame(data)
df_students
```

```python
# Q1:- Calculate total marks and add as new column

df_students["Total Marks"] = df_students["SQL_Marks"] + df_students["Python_Marks"] + df_students["Excel_Marks"]
df_students
```

```python
# Without using .mean

df_students["Average_Marks"] = (
    (df_students["SQL_Marks"] + df_students["Python_Marks"] + df_students["Excel_Marks"]) / 3)

df_students
```

```python
# Q3:- Fetch all students from Tier 1

tier1_students = df_students[df_students["Tier"]=="Tier 1"]
tier1_students
```

```python
# Q5:- Fetch details of student at 2nd index using label

df_students.loc[[2]]
```

```python
# Q5:- Fetch name and average marks of student at 2nd and 4th index using label

df_students.loc[[2,4], ["Name","Average Marks"]]
```

```python
# Q6:- Using index fetch details of student in 2nd row

df_students.iloc[[2]]
```

```python
# Q7:- Using index fetch name and total marks of 1st and 4th student

df_students.iloc[[0,3], [1, 7]]
```

```python
# Q8:- Fetch rows from index 0 to 2 and column index 1 to 4

df_students.iloc[0:3, 1:5]
```

```python
# Q9:- Fetch name and tier for every alternate student with labels

df_students.loc[::2, ["Name","Tier"]]
```

```python
# Q10:- Fetch name and average marks for students in reverse order

df_students.loc[::-1, ["Name","Average Marks"]]
```
